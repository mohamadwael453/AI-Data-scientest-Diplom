# **Learn**

📘 **Object-Oriented Programming (OOP) – Complete Guide with Real-Life Examples**

🔹 **Definition:**
Object-Oriented Programming (OOP) is a programming paradigm that organizes code using **objects**, which contain both **data** (attributes) and **actions** (methods). It helps developers model real-world systems, making programs cleaner, reusable, and easier to manage.

---

🔹 **Why Do We Use OOP?**

* To organize complex programs into manageable pieces
* To model real-world systems naturally
* To enable **code reusability** and scalability
* To improve collaboration in team projects

**Real-life analogy:** Managing real-world items like **cars, smartphones, or bank accounts** — each has its own data and actions, and OOP helps replicate this structure in code.

---

🔹 **Core Principles of OOP (With Real-Life Examples):**

1. **Encapsulation**

* **Definition:** Combining data and actions into one unit and controlling access to the data.
* **Example:** A **smartphone** stores your contacts and messages. You can make a call without knowing the internal circuitry.
* **Why it matters:** Protects sensitive information and keeps code organized.

---

2. **Inheritance**

* **Definition:** Creating a new class based on an existing one, inheriting its features.
* **Example:** A **Car** class defines wheels and engine. **ElectricCar** inherits these features and adds a battery.
* **Why it matters:** Reuses code, avoids duplication, and speeds up development.

---

3. **Polymorphism**

* **Definition:** The ability of one action to take many forms depending on the object.
* **Example:** **“Play music”** behaves differently on devices: smartphone uses Spotify, radio uses FM signals.
* **Why it matters:** Adds flexibility and adaptability to programs.

---

4. **Abstraction**

* **Definition:** Hiding unnecessary details and showing only essential features.
* **Example:** A **TV remote** lets you change channels and volume without knowing the internal electronics.
* **Why it matters:** Simplifies usage and reduces complexity.

---

🔹 **Advantages of OOP:**
✅ Organizes code in a structured way
✅ Promotes **code reuse** through inheritance
✅ Easier to maintain and update programs
✅ Models real-world systems naturally
✅ Supports large-scale and collaborative projects

---

🔹 **Why OOP Matters:**
OOP mirrors real-life objects and their interactions, making software easier to **understand, extend, and maintain**. It is the foundation of modern programming in software, apps, games, automation, and embedded systems.

---

🔹 **Summary:**
OOP principles — **Encapsulation, Inheritance, Polymorphism, Abstraction** — are essential for building **clean, reusable, and scalable software**. Using these principles ensures programs are efficient, safe, and closer to real-world logic.

```python
from abc import ABC , abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

  @abstractmethod
  def perimeter(self):
    pass
```

```python
class Rectangle(Shape):
  def __init__(self,length,width):
    self.length = length
    self.width = width

  def area(self):
     return self.length * self.width

  def perimeter(self):
    return 2 * (self.length + self.width)

r1 = Rectangle(5,10)
print(r1.area())
print(r1.perimeter())
```

```python
class BaseAnimal:
    def __init__(self, name):
        self.name = name

    def sleep(self):
        return f"{self.name} is sleeping."

class Lion(BaseAnimal):
    def roar(self):
        return "ROAR!"

simba = Lion("Simba")
print(simba.sleep())
print(simba.roar())
```

```python
class A:

    def add(self):
        print("Add from class A")

class B(A):

    def sub(self):
        print("Sub from class B")

class C(B):

    def show(self):
        print("Show from class C")

z = C()
z.show()
z.sub()
z.add()
```

```python
# Grandparent
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

# Parent
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

# Child
class ElectricCar(Car):
    def __init__(self, brand, model, battery):
        super().__init__(brand, model)
        self.battery = battery

    def details(self):
        print(f"{self.brand} {self.model} with {self.battery}kWh battery")

tesla = ElectricCar("Tesla", "Model S", 100)
tesla.details()
```

```python
class Engine :
  def start(self):
    print("Engine started")

  def stop(self):
    print("Engine stopped")

class Radio :
  def play_music(self):
    print("Playing music")
class Samart_car (Engine ,Radio ):
  def drive(self):
    print("Car is driving")

my_car = Samart_car()
my_car.start()
my_car.play_music()
my_car.drive()
my_car.stop()
```

```python
class CredirtCard:
  def pay (self , amount):
    print(f"Paid {amount} using credit card")

class DepitCard :
  def pay (self , amount ):
    print (f"Paid {amount} using Depit Card")

def process_payment(payment_method, amount):
    # The function doesn't care which object it is, as long as it has a .pay() method
    payment_method.pay(amount)

process_payment(CredirtCard(), 100)
process_payment(DepitCard(), 50)
```
