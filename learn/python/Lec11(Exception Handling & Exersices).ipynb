Write function take string, this string is brackets and check if the this brackets is balanced

```python
def is_balanced(string):
    """Checks if the brackets in a given string are balanced.

    This function uses a stack-based approach to determine if all opening brackets
    have corresponding closing brackets in the correct order.

    Args:
        string (str): The input string containing brackets.

    Returns:
        bool: True if the brackets are balanced, False otherwise.
    """
    stack = []
    mapping = {')': '(', ']': '[', '}': '{'}
    for char in string:

      if char in ['(', '[', '{']:
            stack.append(char)
      elif char in [')', ']', '}']:
            if not stack: # Corrected from 'if stack is False:'
                return False  # Mismatched closing bracket or no opening bracket
            top_element = stack.pop()
            if mapping[char] != top_element:
                return False  # Mismatched bracket type
    return not stack # If stack is empty, all brackets are balanced

print(f"'()': {is_balanced('()')}")
print(f"'()[]{{}}': {is_balanced('()[]{}')}")
print(f"'(]': {is_balanced('(]')}")
print(f"'([{{}}])': {is_balanced('([{}])')}")
print(f"'((()))': {is_balanced('((()))')}")
print(f"')(': {is_balanced(')(')}")
print(f"'({[]})': {is_balanced('({[]})')}")
```

Write function that return the first character in a string does not repeat

```python
def first_not_repeating_char(string):
  if not string: # Handle empty string case
    return None

  char_counts = {} # Use a dictionary to store character counts
  for char in string:
    char_counts[char] = char_counts.get(char, 0) + 1

  for char in string:
    if char_counts[char] == 1:
      return char # Return the first character found with a count of 1

  return None  # If loop finishes, no such char found

user_input = input("Please enter your characters: ")
result = first_not_repeating_char(user_input)
print("First non-repeating character:", result)
```

write function that check if the year is leap year


```python
def lepear_year(year):
  if year %4 == 0 and year %100 != 0 or year % 400 == 0:
    print(f"{year} is a leap year")
  else:
    print(f"{year} is not a leap year")


lepear_year(int(input("Enter the year : ")))

```
