```python
x= int (input("please enter frist number: "))
y= float (input("please enter second number: "))
op = input("please enter operator: ")
match op :
    case "+" :
        print(x+y)
    case "-" :
        print(x-y)
    case "*" :
        print(x*y)
    case "/" :
        print(x/y)
    case "%" :
        print(x%y)
    case "**" :
        print(x**y)
    case "//" :
        print(x//y)
    case _:
      print("invalid operator")
```
