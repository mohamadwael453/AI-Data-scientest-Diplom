📘 Object-Oriented Programming (OOP) – Complete Overview

🔹 **Definition:**
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of *objects*, where each object contains data (attributes) and behavior (methods). It helps developers structure code in a clean, organized, and reusable way.

---

🔹 **Why Do We Use OOP?**
OOP is used to make programs easier to understand, manage, and scale.
Instead of writing long, unorganized code, we divide the program into smaller parts (objects), each responsible for a specific task.

---

🔹 **When Do We Use OOP?**

* When building large or complex applications
* When code needs to be reusable
* When working on real-world systems (e.g., apps, games, automation systems)
* When multiple developers are working on the same project

---

🔹 **Core Concepts of OOP:**

1. **Encapsulation**
   Combining data and methods in one unit and protecting the data.

2. **Inheritance**
   Reusing code by creating new classes from existing ones.

3. **Polymorphism**
   Using the same method name with different behaviors.

4. **Abstraction**
   Hiding complexity and showing only essential features.

---

🔹 **Basic Elements:**

* **Class** → Blueprint for creating objects
* **Object** → Instance of a class
* **Attributes** → Variables (data)
* **Methods** → Functions (behavior)

---

🔹 **Advantages of OOP:**
✅ Organized and clean code
✅ Code reusability
✅ Easy maintenance and updates
✅ Scalable for large systems
✅ Closer to real-world modeling

---

🔹 **Disadvantages of OOP:**
❌ Harder for beginners to understand
❌ Requires more planning
❌ Can increase complexity in small projects
❌ Slight performance overhead

---

🔹 **Summary:**
OOP is a powerful way to build structured, reusable, and scalable software. It is widely used in modern programming and is essential for any developer.

📘 OOP Basic Elements Explained

🔹 **Basic Elements of OOP:**

1. **Class**
   A class is a blueprint used to create objects. It defines what data and behavior the objects will have.
   👉 Example: A `Car` class defines properties like brand and speed.

---

2. **Object**
   An object is an instance of a class. It represents a real example created from the class.
   👉 Example: A specific car like "BMW" is an object of the Car class.

---

3. **Attributes (Variables)**
   Attributes are variables inside a class that store data about the object.
   👉 Example: brand = "BMW", speed = 120

---

4. **Methods (Functions)**
   Methods are functions inside a class that define the behavior of the object.
   👉 Example: drive() makes the car move.

---

🔹 Explanation of the Example:

* `Car` → Class (blueprint)
* `car1` → Object (real instance)
* `brand`, `speed` → Attributes (data)
* `drive()` → Method (behavior)

---

🔹 Summary:
These basic elements are the foundation of OOP. Understanding them helps you build structured and real-world-like programs.

 ## 🧠 Class Method vs Static Method (Conceptual Explanation)
Understanding the difference between **Class Methods** and **Static Methods** is essential in Object-Oriented Programming (OOP).

---

### 🔹 Class Method

#### 📌 Definition

A **Class Method** is a method that works with the **class itself**, not with individual objects.
It is used when we want to modify or access data that is shared across all instances.

---

#### 🌍 Real-Life Example

Imagine a **University System**:

* The university sets a tuition fee for all students.
* If the university decides to increase the fee, the change affects **every student**.

👉 This action applies to the **entire class**, not just one student.

✔ This is similar to a **Class Method**

---

### 🔹 Static Method

#### 📌 Definition

A **Static Method** is a method that does **not depend on the class or the object**.
It is simply a utility function grouped inside the class.

---

### 🌍 Real-Life Example

Think of a **GPA Calculator**:

* It takes grades as input
* Returns GPA as output

👉 It does not depend on:

* A specific student
* The university system

✔ It’s just a tool

✔ This is similar to a **Static Method**

---

### 🔥 Key Differences

| Feature           | Class Method                      | Static Method    |
| ----------------- | --------------------------------- | ---------------- |
| Depends on        | Class                             | Nothing          |
| Purpose           | Modify shared data                | Utility / helper |
| Real-life example | Changing tuition for all students | GPA calculator   |

---

### 🧠 Summary

* **Class Method** → Works on the whole system (shared data)
* **Static Method** → Works independently as a helper tool



```python
class Car :
  wheels = 4
  def __init__(self,brand,model ,year):
    self.brand = brand
    self.model = model
    self.year = year

  def Display(self) :
    print(f"The car is  a {self.brand} {self.model} {self.year}")


  @classmethod
  def change_wheels(cls,wheels):
    cls.wheels = wheels
    print(f"The number of wheels is {cls.wheels}")

  @staticmethod
  def is_old(year):
    if year > 2020 :
      print("The car is old")
    else :
     print("The car is modern model")


c1 = Car("BMW","M7")
```
