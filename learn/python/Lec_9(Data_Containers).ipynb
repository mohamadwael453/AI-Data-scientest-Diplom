```python
def second_great_number(l1) :
  for i in range(len(l1)) :
    for j in range(len(l1)-1) :
     if l1[j] < l1[j + 1]:
      temp = l1[j]
      l1[j] = l1[j + 1]
      l1[j + 1] = temp
  return l1[1]


x = second_great_number([9,8,6,5,4,3,2])
print(x)
```

```python
def second_largest(l) :
  largest =l[0]
  for i in l :
    if i > largest:
      largest =i
  second = None
  for i in l :
    if i != largest :
      if  second ==None or i >second  :
        second=i
  return second


print(second_largest([2,6,15,19,8,3]))
```
