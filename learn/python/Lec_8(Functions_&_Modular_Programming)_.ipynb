```python
num=567
revers =0
while num > 0:
    digit = num % 10
    revers = revers * 10 + digit
    num //= 10
print(revers)
```

```python
num=456
temp =num
revers =0
while num > 0:
    digit = num % 10
    revers = revers * 10 + digit
    num //= 10
print(revers)
if temp == revers:
    print("palindrome")
else:
    print("not palindrome")
```

```python
def factorial(n):
    if n == 1 or n == 0 :
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(int(input("please enter the number "))))
```

```python
def count_down(n):
  if n==1 :
    return 1
  else :
    print(n)
    return count_down(n-1)
count_down(5)
```

```python
def subsum(x,y) :
    return x+y
def mul(x,y) :
    return x*y
def div(x,y) :
    return x/y
def app(function, x, y ):
  return(function(x,y))

print(app(subsum , 100,33))
print(app(mul ,7,8))
```

```python
num = int(input("Enter a number: "))
string = []

while num > 0:
    digit = num % 2
    string.append(str(digit))
    num = num // 2

string.reverse()  # because binary is built backwards
print("the binary number : ",end="")
print("".join(string))
```
