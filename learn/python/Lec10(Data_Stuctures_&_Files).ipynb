# Learn

🚀 **Arrays (Lists) — The Foundation of Data Structures**

🔹 **Definition:**
An **Array (List in Python)** is a data structure that stores elements in a **contiguous block of memory**, allowing fast access using indices.

---

📌 **Key Idea:**
Each element has an index → direct access.

---

⚡ **Time Complexity:**

* Access → O(1) ✅
* Search → O(n)
* Insert (end) → O(1) amortized
* Insert/Delete (middle) → O(n) ❌

---

✅ **Advantages:**

* Fast access by index
* Simple and widely used

❌ **Disadvantages:**

* Slow insertion/deletion in middle
* Fixed size (in low-level languages)

---

🎯 **Use Cases:**

* Storing sensor data
* Iteration-heavy tasks
* Base structure for many algorithms

---

💡 **Insight:**
Arrays are fast because of **direct memory access**, not because of searching.




#############################################

📚 **Stack — Last In, First Out (LIFO)**

🔹 **Definition:**
A **Stack** is a linear data structure that follows the **LIFO principle**:
👉 Last element added is the first removed.

---

📌 **Operations:**

* Push → add element
* Pop → remove element
* Peek → view top element

---

⚡ **Time Complexity:**

* Push → O(1) ✅
* Pop → O(1) ✅
* Peek → O(1)

---

🎯 **Use Cases:**

* Function calls (call stack)
* Undo/Redo operations
* Expression evaluation

---

💡 **Insight:**
Stacks are essential for **tracking states and backtracking**.




#############################################

🚶 **Queue — First In, First Out (FIFO)**

🔹 **Definition:**
A **Queue** is a linear data structure that follows the **FIFO principle**:
👉 First element added is the first removed.

---

📌 **Operations:**

* Enqueue → add element
* Dequeue → remove element
* Peek → view front

---

⚡ **Time Complexity:**

* Enqueue → O(1) ✅
* Dequeue → O(1) ✅

---

🎯 **Use Cases:**

* Task scheduling
* Real-time systems
* Breadth-First Search (BFS)

---

💡 **Insight:**
Queues are perfect for **order-based processing**.



#############################################

🔗 **Linked List — Dynamic Data Structure**

🔹 **Definition:**
A **Linked List** consists of nodes, where each node contains:

* Data
* Pointer to the next node

---

📌 **Key Idea:**
Elements are **not stored contiguously** — they are connected using pointers.

---

⚡ **Time Complexity:**

* Access → O(n) ❌
* Search → O(n)
* Insert (head) → O(1) ✅
* Delete (known node) → O(1) ✅

---

✅ **Advantages:**

* Efficient insert/delete
* Dynamic size

❌ **Disadvantages:**

* Slow access
* Extra memory for pointers

---

🎯 **Use Cases:**

* Frequent insert/delete operations
* Memory-flexible systems

---

💡 **Insight:**
Linked lists trade **fast access** for **flexibility**.




🚀 **Hash Tables — Definition & Core Idea**

🔹 A **Hash Table** is a data structure that stores data as **key → value pairs**, using a **hash function** to map each key to a specific index in an array.

📌 Formula:
`index = hash(key) % table_size`

👉 Instead of searching linearly, we compute the exact position of the data → **instant access**.

⚡ **Time Complexity (Average):**

* Search → O(1)
* Insert → O(1)
* Delete → O(1)

🎯 **Why it matters:**
Hash tables are one of the fastest data structures for lookup operations and are widely used in:

* Python dictionaries (`dict`)
* Databases
* Caching systems

💡 **Key Idea:**
Speed comes from direct indexing, not searching.


#############################################

⚠️ **Hash Tables — What is a Collision?**

🔹 A **collision** occurs when two different keys produce the same index using the hash function.

📌 Example:
`15 % 10 = 5`
`25 % 10 = 5`

👉 Both keys map to index 5 → conflict.

---

🎯 **Why collisions happen:**

* Limited table size
* Large number of possible keys
* Hash function compression

---

💡 **Important Insight:**
Collisions are **unavoidable** — the goal is not to eliminate them, but to **handle them efficiently**.

---

🔥 **Next Step:**
We need strategies to resolve collisions without losing performance.




#############################################

🚀 **Collision Handling — Open Addressing**

🔹 **Idea:**
Instead of storing multiple elements in one index, we **search for another empty slot** in the table.

👉 All elements stay inside the array.

---

⚙️ **How it works:**
If a position is occupied → probe for next available slot.

---

⚡ **Complexity:**

* Average → O(1)
* Worst → O(n)

---

✅ **Advantages:**

* No extra memory needed
* Better cache performance

❌ **Disadvantages:**

* Performance decreases when table gets full
* More complex than chaining

---

💡 **Key Insight:**
Efficiency depends heavily on how probing is done.



#############################################

🔺 **Heap — Priority-Based Structure**

🔹 **Definition:**
A **Heap** is a complete binary tree used to maintain **priority order**.

---

📌 **Types:**

* Min Heap → smallest on top
* Max Heap → largest on top

---

⚡ **Time Complexity:**

* Insert → O(log n)
* Delete → O(log n)
* Peek → O(1) ✅

---

🎯 **Use Cases:**

* Priority queues
* Scheduling systems
* Dijkstra algorithm

---

💡 **Insight:**
Heaps are not for searching — they are for **priority management**.




#############################################

🌳 **Trees — Hierarchical Data Structure**

🔹 **Definition:**
A **Tree** is a hierarchical structure consisting of nodes connected by edges, starting from a root node.

---

📌 **Binary Search Tree (BST):**

* Left < Root < Right

---

⚡ **Time Complexity (BST):**

* Search → O(log n)*
* Insert → O(log n)*
* Delete → O(log n)*

❌ Worst case → O(n)

---

🔥 **Balanced Trees (AVL / Red-Black):**

* Always maintain height → O(log n) ✅

---

🎯 **Use Cases:**

* File systems
* Databases
* Searching in sorted data

---

💡 **Insight:**
Trees provide **ordered data with efficient operations**.



#############################################

🌐 **Graph — Modeling Relationships**

🔹 **Definition:**
A **Graph** consists of:

* Vertices (nodes)
* Edges (connections)

---

📌 **Types:**

* Directed / Undirected
* Weighted / Unweighted

---

⚡ **Complexity (Adjacency List):**

* Add edge → O(1)
* BFS / DFS → O(V + E)

---

🎯 **Use Cases:**

* Social networks
* Routing systems
* AI & pathfinding

---

💡 **Insight:**
Graphs are used to model **real-world connections and networks**.
